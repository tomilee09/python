imaginemos tener dos tablas altamente correlacionadas, no necesariamente hay una implicancia entre ellas (ej: numero de tornados y perliculas de Will smith pueden estar correlacionadas (cuando una sube la otra sube) pero no existe una implicancia entre ellas)

si ambas están altamente relacionadas podemos, en vez de representar sus valores en 2 dimensiones, podemos simplemente mostrarlas en una dimensión, mostrando que cuando un dato sube, el otro también, esto es para lo que sirve el PCA

```python
# creamos nuestros datos relacionados
import numpy as np
import matplotlib.pyplot as plt
import random
x = 3*np.random.rand(200) # esto crea 200 números
y = 20*x + 2*np.random.rand(200)

x = x.reshape(200, 1)
y = y.reshape(200, 1)

xy = np.hstack([x, y]) 

plt.plot(xy[:, 0], xy[:, 1], '.')
plt.show()
```

```python
# ahora que tenemos los datos, los centramos en el origen (0, 0)
xy = xy - np.mean(xy, axis = 0)
plt.plot(xy[:, 0], xy[:, 1], '.')
plt.show()
```
