<h1> Blurring or Smoothing an image</h1>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gamma correction: el nombre viene de la suma de imágenes que vimos antes, la fórmula era "alpha*img1 + beta*img2 + gamma", lo que hace gamma es sumar un valor constante
# a las imágenes, y por tanto aumenta o disminuye el brillo, esto puede servir para saturar (o insaturar si es que la palabra existe) una imagen, y por tanto podemos
# borrar los detalles de la imagen, haciendo un blurring
perro = cv2.imread("perro.webp").astype(np.float32)/255 # eso de astype .../255 es pa tener los valores entre 0 y 1
perro = cv2.cvtColor(perro, cv2.COLOR_BGR2RGB)
plt.imshow(perro)
```

```python
# aplicamos la correccion gamma
# esta tiene la siguiente fórmula: imagenCorregida = imagen^gamma
gamma = 20
perroGamma = np.power(perro, gamma)
plt.imshow(perroGamma)
```
