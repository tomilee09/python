```python
# bitwise es aplicar tablas de verdad a las im√°genes
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creo los circulos y cuadrados
cuadrado = np.zeros((400, 600))
cuadrado[100:300, 200:400] = 255 # dibujo un cuadrado blanco
circulo = np.zeros((400, 600))
circulo = cv2.circle(circulo, (300, 200), 125, (255), -1) # creo un circulo blanco
```

```python
plt.imshow(cuadrado, cmap="gray")
```

```python
plt.imshow(circulo, cmap="gray")
```

```python
# con lo anterior podemos ver los diferentes casos
# AND
# N     = negro     =  0
# B     = blanco    =  1
# N, N = N
# N, B = N
# B, N = N
# B, B = B
# por tanto deberia verse solo donde el circulo y cuadrado se sobreponen
AND = cv2.bitwise_and(circulo, cuadrado)
plt.imshow(AND, cmap="gray") 
```

```python
# OR
# si cualquiera de las zonas es blanca se ver√° blanca
OR = cv2.bitwise_or(circulo, cuadrado)
plt.imshow(OR, cmap="gray")
```

```python
# NOT
# en esta opcion solo tenemos que poner una imagen, y lo que hace es: si el pixel es blanco lo pongo negro, si es negro
# lo pongo blanco (es como la opcion INV)
NOT1 = cv2.bitwise_not(circulo)
NOT2 = cv2.bitwise_not(cuadrado)
plt.imshow(NOT2, cmap="gray")
```

```python
# XOR
# es como el or pero es m√°s adecuado a lo que estamos acostumbrados, debe cumplirse unicamente una de las condiciones
# por tanto ya no se cumple que sea blanco (1) cuando se sobreponen (ambas condiciones se cumplen)
XOR = cv2.bitwise_xor(cuadrado, circulo)
plt.imshow(XOR, cmap="gray")
```
