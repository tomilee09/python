```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
rgb = Image.open("rgb.jpeg")
rgb
```

```python
# la transformo en array pa modificarla
rgbNumpy = np.array(rgb)
rgbNumpy
```

```python
plt.imshow(rgbNumpy)
```

```python
# Only red channel
onlyRed = rgbNumpy.copy()
onlyRed[:, :, 1] = 0
onlyRed[:, :, 2] = 0
plt.imshow(onlyRed)
```

```python
# Only green channel
onlyGreen = rgbNumpy.copy()
onlyGreen[:, :, 0] = 0
onlyGreen[:, :, 2] = 0
plt.imshow(onlyGreen)
```

```python
# Only blue channel
onlyBlue = rgbNumpy.copy()
onlyBlue[:, :, 0] = 0
onlyBlue[:, :, 1] = 0
plt.imshow(onlyBlue)
```
