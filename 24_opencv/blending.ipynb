```python
# blending es sumar dos imágenes de una forma específica, se hace de la siguiente forma
# newPixel = a*pixel1 + b*pixel2 + c
# es decir:
# newImage = a*image1 + b*image2 + c
# donde a, b, c son enteros pertenecientes a los reales
# a, b, c los llamaremos alpha, beta, gamma
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# leo las imagenes 
perro = cv2.imread("perro.webp")
gato = cv2.imread("gato.jpg")
# las paso de BGR a RGB
perro = cv2.cvtColor(perro, cv2.COLOR_BGR2RGB)
gato = cv2.cvtColor(gato, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(perro)
```

```python
plt.imshow(gato)
```

```python
# notemos que sus tamaños son diferentes, asi que las tenemos que cambiar de tamaño
perro = cv2.resize(perro, (1200, 900))
plt.imshow(perro)
```

```python
# lo mismo pal gato
gato = cv2.resize(gato, (1200, 900))
plt.imshow(gato)
```

```python
# ahora hacemos la suma
blended = cv2.addWeighted(src1=perro, alpha=0.5, src2=gato, beta=0.5, gamma=0)
plt.imshow(blended)
```

```python
# si no queremos cambiar los tamaños, tenemos las siguientes opciones
# sobreponer una imagen chica sobre otra grande 
perroPlusGatito = perro.copy()
perroPlusGatito[300:800, 300:800]= gato[300:800, 300:800]
plt.imshow(perroPlusGatito)
```

```python
# y tenemos la opcion de sumar imágenes de diferentes tamaños
# ES MUY COMPLEJO, SI QUIERES HACERLO VER CLASE 4 CAP3 (BLENDING AND PASTING IMAGES PART TWO) DEL CURSO DE OPENCV EN INGLES
```

```python
# tambien tenemos otras opciones para sumar o restar imágnes
# add
suma = cv2.add(perro, gato)
plt.imshow(suma)
```

```python
# lo anterior es un caso específico de blended, es equivalente a lo siguiente
blendedSuma = cv2.addWeighted(src1=perro, alpha=1, src2=gato, beta=1, gamma=0)
plt.imshow(blendedSuma)
```

```python
# resta
resta1 = cv2.subtract(perro, gato) # hace la resta pero si llega a un número negativo simplemente toma el 0
resta2 = cv2.absdiff(perro, gato) # hace la resta pero toma el valor absoluto
plt.imshow(resta1)
```
