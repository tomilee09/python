```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("rgb.jpeg")
plt.imshow(img)
```

```python
# transformo la imagen de bgr a rgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# podemos cambiar a muchos espacios de color
imgHsl=cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(imgHsl)
```

```python
# podemos cambiar a muchos espacios de color
imgHsv=cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(imgHsv)
```
