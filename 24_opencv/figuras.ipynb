```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("rgb.jpeg")
plt.imshow(img)
```

```python
# ahora si hacemos las figuritas en la imagen, serán simples
cv2.rectangle(img, pt1=(50, 50), pt2=(700, 700), color=(255, 0, 0), thickness=10)
plt.imshow(img)
```

```python
#se puede hacer circulos
cv2.circle(img, center=(800, 400), radius=200, color=(0, 0, 255), thickness=10)
cv2.circle(img, center=(400, 400), radius=200, color=(0, 0, 255), thickness=-1) # el -1 hace que quede completamente pintado
# lo anterior aplica pa los cuadrados
plt.imshow(img)
```

```python
# se pueden hacer lineas
cv2.line(img, pt1=(0, 0), pt2=(1400, 800), color=(123, 234, 41), thickness=5)
plt.imshow(img)
```

```python
# podemos poner texto
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, text="hello", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=3)
plt.imshow(img)
```

```python
# podemos crear un poligono de cualquier forma
pts= np.array([[100, 300], [200, 200], [400, 300], [200, 400]])
pts= pts.reshape(-1, 1, 2) # estp es obligatorio, por algun motivo opencv lo pide
cv2.polylines(img, [pts], isClosed=True, color=(100, 100, 200), thickness=20)
plt.imshow(img)
```
