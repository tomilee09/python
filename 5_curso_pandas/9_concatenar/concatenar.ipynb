```python
import pandas as pd
import numpy as np
```

```python
# creo las listas, nparrays y df's
lista1 = [i for i in range(0, 10)]
lista2 = [i for i in range(10, 20)]
nparray1 = np.array(lista1)
nparray2 = np.array(lista2)
df1 = pd.DataFrame(lista1)
df2 = pd.DataFrame(lista2)
```

```python
# concatenar lista
lista_suprema = lista1 + lista2
lista_suprema
```

```python
# concatenar npArray
npArray_supremo = np.concatenate([nparray1, nparray2], axis = 0) # axis = 0 est√° por defercto pero lo puse pa que se sepa de la existencia de axis = 1
npArray_supremo
```

```python
concat = pd.concat([df1, df2], axis = 1) # axis = 1 las guarda una columna al lado de la otra
concat
```

```python
# aparece ese indice feo, asi que voy a eliminar los indices
concat.reset_index(drop=True)
```
