```python
import pandas as pd
import numpy as np
import seaborn as sns
df = sns.load_dataset('diamonds')
df
```

```python
# tengo caracteristicas importantes mediante las cuales puedo agrupar, por ejemplo 'cut'
# muestro la media de los datos agrupados por cut
df.groupby('cut').mean()
```

```python
# muestro la mediana de los datos agrupados por cut
df.groupby('cut').median()
```

```python

# muestro la cantidad de los datos agrupados por cut
df.groupby('cut').count()
# y asi con muchas funciones, ej: .max(), .min()
```

```python
# puedo agrupar tomando varios parametros (analogo a usar multiñles indices)
df.groupby(['cut', 'color']).mean()
```

```python
# y puedo pedir que me muestre varias medidas estadisticas con .aggregate(['...'])
df.groupby(['cut', 'color']).aggregate([min, np.mean])
```

```python
# además podemos usar .aggregate con un diccionario que contenga específicamente lo que queremos para cada columna
diccionario = {'carat':[min, max],
                'price':[np.mean]}
df.groupby(['cut', 'color']).aggregate(diccionario)
```

```python
# Además podemos crear filtros con .filter()
def miFiltro(x):
    return np.mean(x['price']) > 4
df.groupby('cut').filter(miFiltro)
```
