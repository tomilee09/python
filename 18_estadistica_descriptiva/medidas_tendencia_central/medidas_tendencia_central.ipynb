```python
sns.displot(df, x = 'price_usd', hue = 'manufacturer_name')
```

ese gráfico está muy cargado, así que vamos a pasar a otra visualización más sencilla, vamos a disminuir las categorias

```python
sns.displot(df, x = "price_usd", hue = "engine_type", multiple = "stack")
```

```python
# notamos que hay muy pocos autos eléctricos, comparemoslos
df.groupby('engine_type').count()
```

```python
# si queremos inspeccionar una marca en particular podemos hacer lo siguiente
Q7 = df[(df["manufacturer_name"] == "Audi") & (df["model_name"] =="Q7")]
sns.displot(Q7, x = 'price_usd', hue= "year_produced")
```
