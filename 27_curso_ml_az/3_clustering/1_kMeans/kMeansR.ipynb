# K-Means

```r
df = read.csv("../Mall_Customers.csv")
x = df[, 4:5]
```

```r
# aplico el método del codo
# kmeans ya esta disponible de base
set.seed(0)
wcss = vector()
for (i in 1:10){
    resultado_kmeans = kmeans(x, i) # se hace el kmeans con los datos x e i numero de clusters
    # al i-esimo elemento de wcss le asigno el valor de la suma de los withinss = within square sum
    wcss[i] <- sum(resultado_kmeans$withinss)
}
wcss
```

```r
# graficamos el método del codo
plot(1:10, wcss, type="b")
```

podemos ver que despues de 5 no baja mucho más, asi que el número optimo de clusters es 5

```r
# ahora aplico el kmeans con 5 clusters, que es el que voy a usar
set.seed(29)
kmeans5 <- kmeans(x, 5, iter.max = 300, nstart = 10) # no entiendo bien nstart, significa cuantos centros se eligen pa ver si convergen?
```
