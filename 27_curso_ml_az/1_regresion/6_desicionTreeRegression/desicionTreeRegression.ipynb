# Arboles de desici칩n usados para calcular regresiones lineales

advertencia: este es un algoritmo discreto

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
```

```python
x = df.iloc[:, 1:2]
y = df.iloc[:, -1]
```

```python
# importo la funcion que me sirve pa hacer la regresion con arboles de desicion
from sklearn.tree import DecisionTreeRegressor
```

```python
regresion = DecisionTreeRegressor(random_state=0) # pa que coincida con el del que ense침a
regresion.fit(x, y)
```

```python
# grafico los resultados
plt.scatter(x, y)
plt.plot(x, regresion.predict(x))
plt.show()
```

```python
# grafico pero ahora con m치s presici칩n
x = x.values # pa transformarlo en np.array y poder usar min() y max()
rango = np.arange(min(x), max(x), 0.1)
rango = rango.reshape(-1, 1) # porque me lo pide xD
plt.scatter(x, y)
plt.plot(rango, regresion.predict(rango))
plt.show()
```

como vemos, hay un problema de overfiting, el arbol de desicion ha separado cada uno de los puntos en vez de crear algunos grupos, esto lo resolveremos luego con los bosques
