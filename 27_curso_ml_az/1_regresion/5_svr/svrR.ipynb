# Support vector machines (SVMs)

```r
df = read.csv("Position_Salaries.csv")
df = df[2:3] # no uso el texto
```

```r
# llamo a esta libreria (se llama asi por un grupo de investigacion que se llama asi)
# install.packages("e1071")
library(e1071)
```

```r
regresion = svm(formula=Salary~.,
                data = df,
                type = "eps-regression", # type especifica el tipo de problema, en este caso de regresion eps
                kernel = "radial") # esto hay que usarlo porque nuestro problema no es lineal (es exponencial)
summary(regresion)
```

```r
require(ggplot2)
```

```r
ggplot() +
    geom_point(aes(x = df$Level, y = df$Salary)) +
    geom_line(aes(x = df$Level, y = predict(regresion, newdata = df))) 
    #theme(text=element_text(family="Mono")) # no me funcionaban las letras, esto solo cambia la font family
```
