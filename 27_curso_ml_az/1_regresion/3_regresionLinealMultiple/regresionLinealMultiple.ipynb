<h1> Regresion Lineal Multiple </h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./50_Startups.csv")
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# tengo distintos estados, california, new york y florida, asi que hay que separarlos en variables dummies
# x = pd.DataFrame(x)
x = pd.concat((x,pd.get_dummies(x.iloc[:, 3])),1)
x = x.drop([x.columns[3]], axis=1)
x
```

```python
# pero aqui hay colinaridad, porque, si ya sabemos que una empresa no está en california ni en florida, entonces estará en new york,
# es decir, NY = 1-CA-FL, es decir, new york es una variable que depende de las 2 otras variables, así que la vamos a eliminar
# (podriamos haber eliminado cualquiera)
#?pd.DataFrame.drop
x = x.drop(columns=["New York"])
x
```

```python
# separo entre train y test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

<h2> Regresion Lineal

```python
from sklearn.linear_model import LinearRegression
regresion = LinearRegression()
regresion.fit(x_train, y_train)
```

```python
# prediccion de los datos
y_pred = regresion.predict(x_test)
```

```python
# parecen ser muy buenas predicciones :D
print((y_test.values-y_pred)/y_test)
```

## creacion del mejor modelo lineal multiple con el método de la eliminación hacia atrás

El modelo es del tipo
$$y = x_0 + x_1 *a_1 + x_2*a_2 + ... + x_n * a_n$$
y queremos saber cual de los coeficientes $x_n$ no es importante, es decir, cuando $a_n \approx 0$  
tambien queremos saber si $x_0 \approx 0$, asi que vamos a agregar una columna de 1's 

```python
import statsmodels.api as sm
```

```python
# agrego la columna de 1's
x.insert(loc=0, column="a_0", value=np.ones(df.shape[0]))
#df["a_0"] = np.ones(df.shape[0])
x
```

```python
# creo la variable donde estan todas las variables, luego se irán dejando solo las optimas
variablesOptimas = x
regresionLineal = sm.OLS(y, variablesOptimas).fit()
regresionLineal.summary()
```

Hay que fijarse en el p_valor, mientras las bajo sea el p_valor, más importante es la variable, pues significa que no se sabe muy bien por donde irá la probabilidad de que algo ocurra, y por tanto, va a modificar más los datos  
El p_valor mide la probabilidad de que azarozamente un valor caiga en el 95% de los datos, mientras más facilmente caiga ahí, mayor será el p_valor
por tanto, si el p valor es bajo, entonces significa que esa variable ... ?

### Lo importante, eliminamos las que tienen p_valor más alto
(hasta que p_valor (P>|t|) sea menor a 0.05)
