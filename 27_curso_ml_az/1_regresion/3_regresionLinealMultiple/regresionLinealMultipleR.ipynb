<h1> Regresión Lineal Múltiple

```r
df <- read.csv("50_Startups.csv")
df
```

```r
# convierto en 3 variables la variable categórica "estados"
df$State = factor(df$State,
                levels = c("New York", "California", "Florida"),
                labels = c(1, 2, 3))
# asi no se hace en python, creo que lo hacemos de esta forma aqui porque en el siguiente paso el algoritmo identifica estas variables
# como variables dummy
```

```r
# dividir entre entrenamiento y testing
library(caTools)
set.seed(123)
split = sample.split(df$Profit, SplitRatio=0.2)
trainSet = subset(df, split==FALSE)
testSet = subset(df, split==TRUE)
trainSet
```

<h2> Crear modelo regresión lineal múltiple

```r
# esto significa "crea la regresion lineal con Profit como variable dependiente y todas las demás como variables independientes"
regresion = lm(formula=Profit ~ .,
            data = trainSet)
summary(regresion)
```

Aqui en la parte de coeficientes R nos indica con <strong>estrellitas *** </strong> cuales son las variables importantes (nos spoilea que son solo R.D spend y el intercepto)

<h2> hacemos las predicciones </h2>

```r
y_pred = predict(regresion, newdata = testSet)
# buenas predicciones :D
(y_pred - testSet[, 5])/testSet[, 5]
```

<h2> Creación de un modelo óptimo </h2>

```r
# hacemos la eliminación hacia atrás
# empezamos con todas las variables y vamos borrando
regresion = lm(formula=Profit ~ .,
            data = trainSet)
summary(regresion)
```

### Ya hice la eliminación hacia atrás en python, al final lo mejor es dejar el gasto en i+d (investigacion (R.D.)) y marketing

```r
regresion = lm(formula=Profit ~ R.D.Spend + Marketing.Spend,
            data = trainSet)
summary(regresion)
```
