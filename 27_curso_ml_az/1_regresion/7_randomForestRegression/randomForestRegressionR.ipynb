# Regresion con random forest


```r
df = read.csv("Position_Salaries.csv")
df = df[2:3] # no uso el texto
df
```

```r
# paquete que sirve pa usar los random forest
require(randomForest)
```

```r
set.seed(42)
regresion = randomForest(x = df[1], 
                        y = df$Salary,
                        ntree = 10)
```

```r
library(ggplot2)
```

```r
rango = seq(min(df$Level), max(df$Level), 0.01)
df_rango = data.frame(Level = rango)
ggplot() +
    geom_point(aes(x = df$Level, y = df$Salary)) +
    geom_line(aes(x = rango, y = predict(regresion, newdata = df_rango))) 
```

```r
# ahora que pasa si aumentamos los arboles
regresion = randomForest(x = df[1], 
                        y = df$Salary,
                        ntree = 500)
```

```r
rango = seq(min(df$Level), max(df$Level), 0.01)
df_rango = data.frame(Level = rango)
ggplot() +
    geom_point(aes(x = df$Level, y = df$Salary)) +
    geom_line(aes(x = rango, y = predict(regresion, newdata = df_rango))) 
```
